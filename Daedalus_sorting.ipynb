```python
import os
import glob
import pandas as pd
```

```python
os.getcwd()
```

```python
a = glob.glob('team_backup\*')
```

```python
good_alphas = []
columns = []

for filename in a:
    #print(filename)
    df = pd.read_csv(filename)
    columns = df.columns
    df = df[(df['is.sharpe']>0.8) & (df['train.sharpe']>0.8) & (df['test.sharpe']>0.8)]
    good_alphas = good_alphas + df.values.tolist()

len(good_alphas)
```

```python
good_alphas_df = pd.DataFrame(good_alphas, columns = columns)
good_alphas_df = good_alphas_df.sort_values(by=['train.sharpe'], ascending=False)
```

```python
save_df = good_alphas_df[['id', 'regular.code', 'is.pnl', 'is.margin', 'is.fitness', 'is.sharpe', 'train.pnl', 'train.margin', 'train.fitness', 'train.sharpe']].sort_values(by=['train.sharpe'], ascending = False)
save_df
```

```python
save_df.to_csv('Daedalus_sorted.csv')
```
